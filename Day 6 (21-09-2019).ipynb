## Visualizing Data
- an inresting way to display Dataframe
- with matplotlib

```python
import matplotlib.pyplot as plt
a=[1,2,3,4]
b=[5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a,b,color='pink')
plt.ylabel('y-Axis',color='red')
plt.xlabel('x-Axis',color='red')
plt.show()
```

```python
marks=[79,82,75,63]
divisions=['A','B','C','D']
index=np.arange(len(divisions))
plt.barh(divisions,marks,color='aqua')
```

```python
lang=['python','java','c','c++']
rating=[75,70,55,65]
plt.pie(rating,labels=lang,shadow=True,startangle=150,autopct='%2.5f%%')
plt.legend(loc='best')
plt.show()
```
