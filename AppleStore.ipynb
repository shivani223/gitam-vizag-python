### Mini-Project on Appstore Data-set

- 1.Calculate the average rating of Gaming and Non-Gaming apps 
- 2.Draw Pie plot based on labels as  below Price category and values as count of each category apps  
labels:

- Price =0  - Free apps
- Price <20 - Affordable apps
- Price<50 - Expensive apps
- Price >50 - Very expensive

```python
import pandas as pd
import matplotlib.pyplot as plt
filepath ='AppleStore.csv'
def readCsvData(filepath):return pd.read_csv(filepath)
df = readCsvData(filepath)
print(df)
```

```python
t= df[['prime_genre','user_rating']]
g= t[t.prime_genre =='Games']
ng=t[t.prime_genre !='Games']
print("Mean {0}".format(g.mean()))
print("Mean non games{0}".format(ng.mean()))
```

```python
movies=['Free apps','AffordableApps','Expensive apps','Very Expensive']
rating1=df[df.price==0]
rating2=df[(df.price>0) & (df.price<20)]
rating3=df[(df.price>20) & (df.price<50)]
rating4=df[(df.price>50)]
free=rating1.count()[0]
af_ap=rating2.count()[0]
ea= rating3.count()[0]
ve= rating4.count()[0]
li=[free,af_ap,ea,ve]
plt.pie(li,labels=movies,shadow=True,startangle=90)
plt.show()
```
